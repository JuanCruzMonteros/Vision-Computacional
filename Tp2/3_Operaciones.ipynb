```python
def RGB2Gris(IMG):
  RB = IMG[:,:,0]
  GB = IMG[:,:,1]
  BB = IMG[:,:,2]
  #cv2_imshow(BB)
  RGBB = np.concatenate((RB,GB,BB),axis=1)
  RdB = RB.astype(float)
  GdB = GB.astype(float)
  BdB = BB.astype(float)
  kB  = (0.1,0.4,0.5)
  ZdB = kB[0]*RdB+kB[1]*GdB+kB[2]*BdB
  ZB  = ZdB.astype(int)
  return ZB
```

# Lectura de imagen a color

```python
Icv = imread('Montana.jpg')      # lectura en formato opencv (BGR)
Z  = RGB2Gris(Icv)

```
